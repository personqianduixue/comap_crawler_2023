```python
import shutil
import pandas as pd

shutil.copyfile('./all/all.txt', './all/all.csv')
data = pd.read_csv('./all/all.csv')
data.head()
```

```python
univ_data=data['university'].value_counts().iloc[0:20]
univ_data
```

```python

ax=univ_data.sort_values().plot.barh()
fig=ax.get_figure()
fig.savefig('统计结果.png',bbox_inches='tight')
```
